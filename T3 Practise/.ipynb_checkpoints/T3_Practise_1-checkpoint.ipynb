```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('innercity.xlsx')
# df = df.drop(columns=['gender'])
df.head()
```

# Are there any missing values in the dataset? How would you handle them?

```python
df = df.dropna()
# df = df.fillna(df.mean())
df = df[~df.apply(lambda row: row.astype(str).str.contains('$',regex=False)).any(axis=1)]
```

# What is the distribution of house prices in the dataset?

```python
house_price = df['price']

plt.figure(figsize=(10,6))
# plt.hist(house_price)
sns.histplot(house_price, bins=30, kde=True)
plt.title('Distribution of House Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# How do the number of bedrooms and bathrooms vary across different houses?

```python
bedrooms = df['room_bed']
bathrooms = df['room_bath']

fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,6))
sns.histplot(bedrooms,bins=20,ax=axes[0])
axes[0].set_title('Number of Bedrooms')
axes[0].set_xlabel('Bedrooms')
axes[0].set_ylabel('Total Number')

sns.histplot(bathrooms,bins=20,ax = axes[1])
axes[1].set_title('Distribution of Bathrooms')
axes[1].set_xlabel('Bathrooms')
axes[1].set_ylabel('Total Number')

plt.tight_layout()
plt.show()
```

# Is there any correlation between living area size and house prices?

```python
living_area = df['living_measure']
prices = df['price']

corrcoef = np.corrcoef(living_area,prices)[0,1]
plt.figure(figsize=(12,6))
sns.scatterplot(x=living_area,y=prices)
plt.title('Correlation Between Living area and House Prices')
plt.xlabel('Price')
plt.ylabel('Living Area')
plt.text(0.1,0.9, f'Correlation Coefficient: {corrcoef:.2f}', transform=plt.gca().transAxes,fontsize=15,ha='left')
plt.show()
```

# How does the year built or renovated affect the house prices?

```python
year_built = df['yr_built']
year_renovated = df['yr_renovated']
prices = df['price']

fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
sns.lineplot(x=year_built,y=prices,estimator='mean',ci=None,ax=axes[0])
axes[0].set_title('Year Built vs. House Prices')
axes[0].set_xlabel('Year Built')
axes[0].set_ylabel('House Prices')

sns.lineplot(x=year_renovated,estimator='mean',y=prices,ci=None,ax=axes[1])
axes[1].set_title('Year Renovated vs. House Prices')
axes[1].set_xlabel('Year Renovated')
axes[1].set_ylabel('Price')

plt.tight_layout()
plt.show()

# to remove outliers after seeing the chart
min_year_built = df['yr_built'].min()
df = df[df['yr_renovated'] >= min_year_built]

year_built = df['yr_built']
year_renovated = df['yr_renovated']
prices = df['price']

fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
sns.lineplot(x=year_built,y=prices,estimator='mean',ci=None,ax=axes[0])
axes[0].set_title('Year Built vs. House Prices')
axes[0].set_xlabel('Year Built')
axes[0].set_ylabel('House Prices')

sns.lineplot(x=year_renovated,estimator='mean',y=prices,ci=None,ax=axes[1])
axes[1].set_title('Year Renovated vs. House Prices')
axes[1].set_xlabel('Year Renovated')
axes[1].set_ylabel('Price')

plt.tight_layout()
plt.show()
```

# Are there any significant differences in prices based on location (zipcode, latitude, longitude)?

```python
df = df[~df.apply(lambda row: row.astype(str).str.contains('$',regex=False)).any(axis=1)]

zipcode = df['zipcode']
latitude = df['lat']
longitude = df['long']
prices = df['price']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

sns.barplot(x=zipcode, y=prices, ax=axes[0])
axes[0].set_title('House Prices by Zipcode')
axes[0].set_xlabel('Zipcode')
axes[0].set_ylabel('House Prices')

sns.scatterplot(x=latitude, y=prices, ax=axes[1])
axes[1].set_title('House Prices by Latitude')
axes[1].set_xlabel('Latitude')
axes[1].set_ylabel('House Prices')

sns.scatterplot(x=longitude, y=prices, ax=axes[2])
axes[2].set_title('House Prices by Longitude')
axes[2].set_xlabel('Longitude')
axes[2].set_ylabel('House Prices')

plt.tight_layout()
plt.show()
```

# Can you create new features such as total number of rooms or price per square foot?

```python
df['total_rooms'] = df['room_bed'] + df['room_bath']
df['price_per_sqft'] = df['price'] / df['living_measure']

# plt.figure(figsize=(12,8))
sns.histplot(df['total_rooms'],kde=True)
plt.title('Distribution of Total Number of Rooms')
plt.xlabel('Frequency')
plt.ylabel('Total Number of Rooms')
plt.show()

sns.histplot(df['price_per_sqft'],kde=True)
plt.title('Distribution of Price per square foot')
plt.xlabel('Frequency')
plt.ylabel('Price per square foot')
plt.show()
```

# Is there a way to combine the 'living_measure' and 'lot_measure' to create a more meaningful feature?

```python
df['total_measure'] = df['living_measure'] + df['lot_measure']

# plt.figure(figsize=(8, 6))
sns.histplot(df['total_measure'],kde=True)
plt.title('Distribution of Total Measure')
plt.xlabel('Frequency')
plt.ylabel('Total Measure')
plt.text(0.1,0.9,f'Total Measure in Sqre Mtre',transform=plt.gca().transAxes,fontsize=13,ha='left')

plt.show()
```

# Simple Linear Regression

```python
df.head()
```

```python
x = df.iloc[:,:2].join(df.iloc[:,3:])
x
```

```python
y = df.iloc[:,2]
y
```

# Encoding the Independent Variable

```python
categorical_columns = ['gender','dayhours']

# Perform one-hot encoding for categorical columns
X_encoded = pd.get_dummies(x, columns=categorical_columns)
print(X_encoded.head(1))
```

# Splitting the data as Train set & Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_encoded,y, test_size=0.2, random_state=0)
```

# Training Simple Linear regression model using Train set data

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

# Predicting the Test Result

```python
# y_pred = regressor.predict(x_test)
rf_predictions = regressor.predict(x_test)
```

```python
regressor.score(x_test,y_test)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, rf_predictions, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--', color='black')
plt.title('Actual vs. Predicted Prices (Random Forest)')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.grid(True)
plt.show()
```

# Evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2_score = regressor.score(x_test, y_test)
```
